```python
%matplotlib inline

import networkx as nx
import matplotlib.pyplot as plt

# 1. Crear un grafo direccionado (DiGraph)
G = nx.DiGraph()

# 2. Definir tus conexiones
conexiones = [
    ("A", "B"), ("A", "C"), ("B", "C"), ("B", "D"),
    ("C", "D"), ("C", "E"), ("D", "E"), ("D", "F"),
    ("E", "F"), ("E", "G"), ("F", "G"), ("F", "H"),
    ("G", "H"), ("G", "A"), ("H", "A"), ("H", "B")
]

# 3. Añadir las conexiones al grafo
G.add_edges_from(conexiones)

# 4. Dibujar la red
pos = nx.circular_layout(G)
nx.draw(G, pos, with_labels=True, node_color='green',
        node_size=800, font_weight='bold', arrowsize=20, edge_color='violet')

plt.title("Red Direccionada de 8 Nodos (Grado de salida = 2)")
plt.show()
```
